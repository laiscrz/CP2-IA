### *Análise de Agrupamento*

Neste tópico, vamos abordar a análise de agrupamento de variáveis é uma técnica exploratória usada para identificar padrões e relacionamentos entre as variáveis em um conjunto de dados.

```python
# Definindo faixas de faturamento, margem bruta e total de ativos para estimar o tamanho das empresas
faixas_faturamento = [2, 1000000, 5000000, 10000000, 50000000, float('inf')]  # em reais
faixas_margem_bruta = [0, 0.1, 0.3, 0.5, 1]  # porcentagem
faixas_total_ativos = [0, 1000000, 5000000, 10000000, 50000000, float('inf')]  # em reais

rotulos_tamanhos = ['Microempresa', 'Pequena Empresa', 'Média Empresa', 'Grande Empresa', 'Corporação']
```

```python
# Criando novas colunas 'Tamanho da Empresa' com base nas faixas de faturamento, margem bruta e total de ativos definidas
df['Tamanho da Empresa_Faturamento'] = pd.cut(df['faturamentoBruto'], bins=faixas_faturamento, labels=rotulos_tamanhos, right=False)
df['Tamanho da Empresa_Margem'] = pd.cut(df['margemBruta'], bins=faixas_margem_bruta, labels=rotulos_tamanhos[:-1], right=False)
df['Tamanho da Empresa_Ativos'] = pd.cut(df['totalAtivo'], bins=faixas_total_ativos, labels=rotulos_tamanhos, right=False)
```

```python
# Normalizando os dados dos atributos faturamento bruto, margem bruta e total de ativos
scaler = MinMaxScaler()
df_normalizado = df[['faturamentoBruto', 'margemBruta', 'totalAtivo']]
df_normalizado = pd.DataFrame(scaler.fit_transform(df_normalizado), columns=df_normalizado.columns)
```

```python
# Atribuição de Pesos
pesos = {'faturamentoBruto': 0.6, 'margemBruta': 0.3, 'totalAtivo': 0.1}

# Cálculo do Score Composto
df['Score'] = df_normalizado['faturamentoBruto'] * pesos['faturamentoBruto'] + \
              df_normalizado['margemBruta'] * pesos['margemBruta'] + \
              df_normalizado['totalAtivo'] * pesos['totalAtivo']

# Definição dos Grupos de Tamanho de Empresa com base no score composto
faixas_score = [0, 0.3, 0.6, 1]  # Definindo faixas de score
rotulos_tamanhos = ['Pequena Empresa', 'Média Empresa', 'Grande Empresa']
df['Tamanho da Empresa'] = pd.cut(df['Score'], bins=faixas_score, labels=rotulos_tamanhos)
```

```python
# Atualizando a categoria MEI
categorias_tamanho = ['Microempresa', 'Pequena Empresa', 'Média Empresa', 'Grande Empresa', 'Corporação', 'MEI']
df['Tamanho da Empresa'] = df['Tamanho da Empresa'].astype('category')
df['Tamanho da Empresa'] = df['Tamanho da Empresa'].cat.set_categories(categorias_tamanho)
df.loc[df['empresa_MeEppMei'] == True, 'Tamanho da Empresa'] = 'MEI'
```

## ***Modelo de Classificação (modelo 1)***

### *Imports necessários*

```python
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
```

### *Coletando os dados*

```python
# Selecionando colunas de entrada e saida
# Atribuindo a coluna 'class' ao Y
Y = df['Tamanho da empresa']

# Atribuindo as outras colunas ao X
X = df[['maiorAtraso', 'margemBrutaAcumulada']]
```

### *Separação de Treino e Teste*

```python
# Dividindo os dados em conjuntos de treinamento e teste (80% treinamento, 20% teste)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Exibindo o tamanho dos conjuntos de treinamento e teste
print("Tamanho do conjunto de treinamento:", X_train.shape)
print("Tamanho do conjunto de teste:", X_test.shape)
```

### *Treinamento do Modelo*

```python
# Modelo => KNN
# Cria o classificar KNN
k = 5
model_knn = KNeighborsClassifier(n_neighbors=k)

# Cria o modelo de machine learning
model_knn.fit(X_train, Y_train)
```

### *Predições do modelo*

```python
# Fazendo previsões no conjunto de teste
y_pred_knn = model_knn.predict(X_test)

print("Predição: {}".format(y_pred_knn))
```

### *Métricas - Acurácia*

```python
accuracy_knn = accuracy_score(Y_test, y_pred_knn)
print("Acerto médio de classificação: ", accuracy_knn)
```

### *Predições fora das amostras do dataset*

```python
# Criamos um modelo utilizando duas entradas e uma saida, logo temos que passar duas entradas para o modelo faça a predição.
model_knn.predict([[3.3, 3.2]])
```
